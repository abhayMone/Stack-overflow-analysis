```python
%%bigquery
WITH
  stackoverflow_ans_delay AS (
  SELECT
    users.reputation,
    ARRAY_LENGTH(SPLIT(questions.tags,"|")) AS num_tags,
    questions.score AS score,
    EXTRACT(YEAR
    FROM
      questions.creation_date) AS year,
    TIMESTAMP_DIFF(answers.creation_date,questions.creation_date,DAY) AS ans_delay_in_days
  FROM
    `bigquery-public-data.stackoverflow.posts_answers`answers
  JOIN
    `bigquery-public-data.stackoverflow.posts_questions` questions
  ON
    answers.id=questions.accepted_answer_id
  JOIN
    `bigquery-public-data.stackoverflow.users` users
  ON
    questions.owner_user_id=users.id
  WHERE
    accepted_answer_id IS NOT NULL )
SELECT
  * EXCEPT (year)
FROM
  stackoverflow_ans_delay
WHERE
  year BETWEEN 2008
  AND 2017
```
